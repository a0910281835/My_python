## imitate  
$ f(x)=1.2x+0.8+\epsilon $ 

```python
x=np.linspace(0,5,200)
y=1.2*x+0.8+0.5*np.random.rand(200)
```

```python
plt.scatter(x,y)
```

## 分資料訓練和測試

```python
from sklearn.model_selection import train_test_split
```

### 數據分析喜歡的形式
$$[x_{1},x_{2},....]$$
### 轉成這種
$$[[x_{1}],[x_{2}]]$$

```python
x=x.reshape(200,1)
```

```python
y=y.reshape(200,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### 線性回歸 所有函數學習

1. 開函數學習機
2. 訓練
3. 預測

```python
## 1.函數學習機
from sklearn.linear_model import LinearRegression
```

```python
## 開機預熱
reg=LinearRegression()
```

```python
## 2.開始訓練
reg.fit(x_train,y_train)
```

```python
## 3 test
reg.predict([[3]])
```

```python
y_pred=reg.predict(x_test)
```

```python
## in fact x need to reveal
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,c='r') ## 這是比然的 因為是linear regr
```

## 房價預測

```python
from  sklearn.datasets import load_boston
```

```python
boston_dataset=load_boston()
```

```python
print(boston_dataset.DESCR)
```

```python
boston_dataset.feature_names
```

```python
## 前三筆
boston_dataset.data[:3]
```

## 如何將資料轉成pandas 資料型態

```python

boston=pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
```

```python
boston.head()
```

```python
boston["MEDV"]=boston_dataset.target
```

```python
## Ya , i have the data
boston
```

```python
# boston's houses average value
boston.MEDV.mean()
```

```python
## 畫圖套件
import seaborn as sns
```

```python
## bins are boxes
sns.distplot(boston.MEDV,bins=30)
```

## 訓練資料

```python
## take the column from CRIM to LSTAT
boston[boston.columns[:-1]].values
```

```python
X=boston.loc[:,"CRIM":"LSTAT"].values
```

```python
X
```

```python
Y=boston.MEDV.values
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
## 函數學習機
reg=LinearRegression()
```

```python
## train
reg.fit(x_train,y_train)
```

```python
## pred
y_pred=reg.predict(x_test)
```

```python
## this is a tool to decorate your graph
sns.set()
```

### 如何看預測與真實值差別 再輸入是多維情況下

```python
plt.title("the x=y is that our pred equal real and if the scatter is in below of the line .that means we underestimate")
plt.scatter(y_test,y_pred)

plt.plot(np.linspace(0,50,100),np.linspace(0,50,100))
plt.xlabel('test')
plt.ylabel('predict')
```

```python
(y_pred-y_test).std()
```

## 對參數做normalization

```python
## 相關係數
correlation_matrix=boston.corr().round(2)
```

```python
####   這邊介紹你如何使用不同顏色更改
import matplotlib
import matplotlib.colors as col
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
startcolor = '#ff0000'   #红色，读者可以自行修改
midcolor = '#00ff00'     #绿色，读者可以自行修改
endcolor = '#0000ff'          #蓝色，读者可以自行修改
cmap2 = col.LinearSegmentedColormap.from_list('own2',[endcolor,midcolor,startcolor])
# extra arguments are N=256, gamma=1.0
# we can skip name here as it was already defined
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(correlation_matrix,annot=True,cmap=cmap2)
```
