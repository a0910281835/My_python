```python
## basic package
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## SVM package
from sklearn.svm import SVC
```

```python
## load data set of iris
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
```

```python
## 數據資料集 簡介
#print(iris.DESCR)
```

```python
## 輸入 以後面兩個資料來做 petal length and petal width
x=x[:,2:]
```

```python
## 分類機
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=np.random.randint(1,100))
```

```python
## 開空函數機
clf=SVC()
```

```python
## feed the data
clf.fit(x_train,y_train)
```

```python
## predict
y_pred=clf.predict(x_test)
```

## 真實資料來畫圖

```python
##預測準度 只有准與不准
plt.scatter(x_test[:,0],x_test[:,1],c=(y_test-y_pred),cmap='Paired')
```

```python
## 顯示利用此圖應該顯現的圖層 這個圖也意味著他的分群比較準
plt.scatter(x[:,0],x[:,1],c=y,cmap='Paired')
```

## 模擬資料畫圖

```python
## 構建整個整個 [0,8]*[0,3]的圖 以此來看出區分線
x_ar=np.arange(0,8,0.02)
y_ar=np.arange(0,3,0.02)
X,Y=np.meshgrid(x_ar,y_ar)
```

```python
## 融合
P=np.c_[X.ravel(),Y.ravel()]
```

```python
C=clf.predict(P)
```

```python
C
```

```python
## 最終呈現圖  預測圖與真實資料
plt.contourf(X,Y,C.reshape(X.shape)) ## 假資料
plt.scatter(x[:,0],x[:,1],c=y,cmap='Paired')
```
