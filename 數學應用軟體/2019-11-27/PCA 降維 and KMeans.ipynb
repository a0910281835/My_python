```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## PCA 降維 in iris

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
## 壓到二維(類似SVD 找出矩陣 A:R^n -> R^m 最重要的 orthgonal eigenvector)
from sklearn.decomposition import PCA
```

```python
## 開 PCA機器 壓到二維
pca=PCA(n_components=2)
```

```python
## 訓練資料輸入 (從資料中課看出 PCA也算是非監督式學習 因為只餵輸入 沒有為輸出)
pca.fit(x)
```

```python
## 內容轉二維
x[21]
```

```python
## 小 x四維 大 轉二維
X=pca.transform(x)
```

```python
X[21]
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

## 正式訓練資料 SVM

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=np.random)
#x_train
```

```python
## 開函數學習機 SVM (SVR 是用來做回歸)
from sklearn.svm import SVC
clf=SVC(gamma='auto')
```

```python
##  feed data
clf.fit(x_train,y_train)
```

```python
## predicy
y_pred=clf.predict(x_test)
```

## 使用全平面資料來構建 "區分曲線"


```python
x0=np.arange(-4,4.2,0.02)
y0=np.arange(-2,1.6,0.02)
xm,ym=np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()] ## remark :carful to reavel
z=clf.predict(P)
Z=z.reshape(xm.shape) ## countourf
```

```python
## 假定資料(未來預測資料) 與真實資料 圖片
plt.contourf(xm,ym,Z,cmap='Paired',alpha=0.3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='Paired')
```

## 開Mean-shift 分類


```python
from sklearn.cluster import MeanShift
```

```python
## 開空白機 (分群的量多輛少)
clf=MeanShift(bandwidth=0.9)
```

```python
## train data
clf.fit(x)
```

```python
## 顯現結果
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
```

```python
## 顯現 array True table has operation for union and interact
clf.predict(x[((0>x[:,0])&(x[:,0]>-2))&((x[:,1]>-4)&(4>x[:,1]))])
```

```python
## 畫全圖
x0=np.arange(-11,8,0.02)
y0=np.arange(-11,13,0.02)
xm,ym=np.meshgrid(x0,y0)
P=np.c_[xm.ravel(),ym.ravel()]
z=clf.predict(P)
Z=z.reshape(xm.shape)
```

```python
## 太複雜的圖 看起來contourf 無法顯示
plt.contourf(xm,ym,Z,cmap='Paired',alpha=0.3)
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
plt.colorbar()
```

```python
## 等高線 和 點圖
plt.contour(xm,ym,Z)
plt.scatter(xm.ravel(),ym.ravel(),c=z.ravel())
```

```python
## 全點圖
plt.scatter(P[:,0],P[:,1],c=z,cmap='Paired')
plt.scatter(x[:,0],x[:,1],c=clf.labels_,cmap='Paired')
```
