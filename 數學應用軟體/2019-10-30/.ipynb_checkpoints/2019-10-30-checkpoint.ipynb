```python
B=np.array([0.1,0.3,0.4,0.1,0.1])
```

```python
A=np.array(dfs[["國文","英文","數學","自然","社會"]])
```

```python
A
```

```python
## 1way
dfs["權重"]=A@B
```

```python
dfs
```

```python
dfs["權重"]=dfs["數學"]*2+dfs["英文"]*1.5
```

```python
dfs
```

```python
dfs["總分"]=dfs.loc[:,"國文":"自然"].sum(axis=1)
```

```python
dfs
```

```python
dfs=dfs.drop("加權",axis=1)
```

```python
dfs
```

```python
sort_dfs=dfs.sort_values(by='總分',ascending=False)
```

```python
sort_dfs
```

```python
## first consider sum second 權重 
dfs_sorted=dfs.sort_values(by=['總分','權重'],ascending=False)
```

```python
dfs_sorted
```

```python
dfs_sorted.index=range(1,6,1)
```

```python
dfs_sorted
```

```python
dfs_sorted.數學==15
```

```python
dfs_sorted[(dfs_sorted.英文==10)&(dfs_sorted.數學==15)]
```

```python
dfs_sorted[(dfs_sorted.英文==10)|(dfs_sorted.數學==15)]
```

## 刪掉指定行列

```python
## two condition
dfs_sorted.drop(index=(dfs_sorted[(dfs_sorted.英文==10)].index)|(dfs_sorted[(dfs_sorted.數學==8)].index),axis=0)
```

```python
# condition for 國文 <= 10
dfs_sorted.drop(index=dfs_sorted[(dfs_sorted.國文<=10)].index,axis=0)
```
